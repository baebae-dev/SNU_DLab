```python
def impression_lst(x):
    return x.split(' ') 
```

```python
large_tr_behavior.columns = ['Impression ID', 'User ID', 'Impression Time', 'Uesr Click History', 'Impression News']
large_tr_behavior.head() 
```

```python
large_tr_behavior['Impression_News_lst'] = large_tr_behavior['Impression News'].apply(impression_lst)
large_tr_behavior.head()  
```

```python
# 뉴스별 사용자에게 선택된 횟수 
def news_readed_count(impression_news):
    candidate_news = collections.defaultdict(int) 

    for news in tqdm(impression_news):
        for cand_news in news:
            # 사용자가 해당 후보 뉴스 읽었으면
            if int(cand_news.split('-')[1]) == 1:
                candidate_news[cand_news.split('-')[0]] += 1

    return dict(candidate_news)
```

```python
candidate_news = news_readed_count(large_tr_behavior['Impression_News_lst'])
```
